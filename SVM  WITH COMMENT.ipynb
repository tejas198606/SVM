```python
#importing the model
from sklearn.svm import SVC
model=SVC()
```

```python
# fitting the model
model.fit(X,y)
```

```python
# MoDEL SCORE
model.score(X,y)
```

```python
# MODEL PREDICATION
model.predict([[5.1, 3.5, 1.4, 0.2]])
```

```python
# TUNING MODEL
new_model=SVC(C=3)
new_model.fit(X,y)
new_model.score(X,y)
```
